# Logistic Regression with Imbalanced Dataset

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=10000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.99],random_state=10)
```

```python
from collections import Counter
Counter(y)
```

```python
import pandas as pd
pd.DataFrame(X)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
solvers=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
c_values=[100,10,1.0,0.1,0.01]
class_weights=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solvers,class_weight=class_weights)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_pred)
print("accuracy:",score)
print("Confusion Matrix: ",confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Logistic Regression with ROC curve and ROC AUC score

```python
X,y=make_classification(n_samples=1000,n_classes=2,random_state=42)
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
dummy_model_prob=[0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
model_prob=model.predict_proba(X_test)
```

```python
model_prob
```

```python
model_prob=model_prob[:,1]
```

```python
dummy_auc_score=roc_auc_score(y_test,dummy_model_prob)
model_auc_score=roc_auc_score(y_test,model_prob)
print(dummy_auc_score)
print(model_auc_score)
```

```python
dummy_fpr,dummy_tpr,_=roc_curve(y_test,dummy_model_prob)
model_fpr,model_tpr,thresholds=roc_curve(y_test,model_prob)
```

```python
model_tpr
```

```python
from matplotlib import pyplot
pyplot.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy model')
pyplot.plot(model_fpr,model_tpr,linestyle='-',label='Logistic model')
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
pyplot.legend()
pyplot.show()
```

```python
from matplotlib import pyplot
import numpy as np
fig=pyplot.figure(figsize=(20,50))
pyplot.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy model')
pyplot.plot(model_fpr,model_tpr,linestyle='-',label='Logistic model')
ax=fig.add_subplot(111)
for xyz in zip(model_fpr,model_tpr,thresholds):
    ax.annotate('%s' % np.round(xyz[2],2),xy=(xyz[0],xyz[1]))
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
pyplot.legend()
pyplot.show()
```

Select a threshold which gives low false positive rate and high true positive rate, according to the above curve 0.55 is the threshold which gives high tpr and low fpr

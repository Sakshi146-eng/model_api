```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
X=pd.DataFrame(iris['data'],columns=('sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'))
```

```python
iris['DESCR']
```

```python
iris
```

```python
y=pd.DataFrame(iris['target'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
plt.figure(figsize=(20,20))
from sklearn import tree
tree.plot_tree(dtc,filled=True)
```

```python
y_pred=dtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

#Hyperparameter Tunning and Preprunning

```python
params={
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features': ['sqrt', 'log2'],
}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(dtc,param_grid=params,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

# Decision Tree Regressor

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
dataset['DESCR']
```

```python
X=pd.DataFrame(dataset['data'],columns=('age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'))
```

```python
y=pd.DataFrame(dataset['target'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(15,15))
import seaborn as sns
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(X_train,y_train)
```

```python
params={
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['sqrt', 'log2']
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(dtr,param_grid=params,cv=5,scoring='neg_mean_absolute_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
model=DecisionTreeRegressor(criterion= 'poisson',max_depth=3,
 max_features= 'log2',
 splitter= 'best')
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(model,filled=True)
```
